Write a function that gets the most recent share price for your portfolio of stocks, which consists of the largest 5 companies by market capitalization, and returns a dictionary containing the most recent price for each stock.

```python
portfolio = ['MSFT', 'AAPL', 'NVDA', 'GOOGL', 'AMZN']
api_key = 'YOUR API KEY HERE'

def get_portfolio_prices(portfolio, api_key):
    import pandas as pd
    portfolio_prices = {}
    for stock in portfolio:
        url = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol='+stock+'&interval=5min&apikey='+api_key
        data_json = requests.get(url).json()
        most_recent_quote_time = sorted(list(data_json['Time Series (5min)'].keys()), reverse=True)[0]
        portfolio_prices[stock] = data_json['Time Series (5min)'][most_recent_quote_time]['4. close']
    return portfolio_prices
```
